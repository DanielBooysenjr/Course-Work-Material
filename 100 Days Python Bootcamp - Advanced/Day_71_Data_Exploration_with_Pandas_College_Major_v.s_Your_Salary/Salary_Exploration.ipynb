```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Get the HTML content of the webpage
url = "https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors"
response = requests.get(url)

# Step 2: Parse the HTML content with Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Locate the table in the HTML code
table = soup.find('table', class_='data-table')

# Step 4: Extract the column headers from the table
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

# Step 5: Extract the rows from the table
rows = []
for tr in table.find_all('tr'):
    row = []
    for td in tr.find_all('td'):
        row.append(td.text.strip())
    rows.append(row)

# Step 6: Create a Pandas dataframe with the column headers and rows
df = pd.DataFrame(rows, columns=headers)

# Step 7: Manipulate and analyze the data as needed
# ...

# Step 8: Export the dataframe to a CSV file, Excel file, or any other format
final_data = df.to_csv('my_table_data.csv', index=False)

```
